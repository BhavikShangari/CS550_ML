```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Question:**
1. Imagine a machine learning model designed to diagnose a rare disease that affects only 1% of the population. In a test set of 10,000 patients, only 100 actually have the disease. The confusion matrix from the model's predictions might look like this:

    True Positives (TP): 80 patients correctly diagnosed with the disease.

    False Negatives (FN): 20 patients who have the disease but were incorrectly diagnosed as not having it.

    False Positives (FP): 100 patients who do not have the disease but were incorrectly diagnosed as having it.

    True Negatives (TN): 9,800 patients correctly diagnosed as not having the disease.

# so which performance parameter you should prefer for model performance?


```python
a = np.array([[80, 100],
              [20, 9800]])
sns.heatmap(a, annot=True , xticklabels=['positive', 'negative'], yticklabels=['positive', 'negative'])
plt.xlabel("Labels")
plt.ylabel("Predicted")
```

So, we have a rare disease which affects only 1 % population, so the data is imbalanced, and we will have 99% accuracy with the simple model itself.

In our case, the more harmful thing is that if the person has disease and the model says the he/she does not has it, which is False Negative, so we need to make sure that False negatives are mitigated.

Other thing which we need to make sure after making False Negatives sure is the case of False Positives, the reason for that is if the person does not has the disease and model predicts that the person has disease, it will waste the resources and money required for further medical tests or operations. This is not much harmful as the previous one as the life is not at risk.

For evaluating False negatives we have a metric **Recall**

For evaluating False positives we have a metric **Precision**

### The First preference for the metric should be given to **Recall**
### The Second preference for the metric should be given to **Precision**

### But if there is a constraint that we can use only one metric, it would be beneficial to use **F1 Score**, as it combines both Precision and Recall by taking their harmonic means
