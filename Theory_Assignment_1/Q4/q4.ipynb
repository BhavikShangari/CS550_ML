```python
col_idx_min_max = []
for i in range(len(df.columns)):
    if df.iloc[:,i].dtype == np.float64:
        col_idx_min_max.append(i)
df.columns[col_idx_min_max], col_idx_min_max
```

```python
def binarize(value):
    if value <= 6:
        return 0
    else:
        return 1
df['quality'] = df['quality'].map(binarize)
df.quality.unique()
```

```python
def min_max_scaler(df, col_idxs):
    temp = df.iloc[:,col_idxs]
    min_vals = np.min(temp, axis=0).values
    max_vals = np.max(temp, axis=0).values
    
    temp = (temp - min_vals) / (max_vals - min_vals)
    df.iloc[:,col_idxs] = temp
    return df
    
    
min_max_scaler(df, col_idx_min_max).describe() 
```
